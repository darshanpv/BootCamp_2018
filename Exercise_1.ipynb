Find the image difference:

* Create a new pyhton notebook
* Load images chair_1.png and chair_2.png
* Compute the difference between two images (use cv2.absdiff function)
* Display the output image

```python
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

import cv2
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

from pylab import rcParams
rcParams['axes.xmargin'] = 0
rcParams['axes.ymargin'] = 0
rcParams['figure.figsize'] = (14,6)
```

## Find Image difference

```python
img1 = cv2.imread('images/chair_1.png')
img2 = cv2.imread('images/chair_2.png')
# check image shape and channel
print("Image -1 Shape: ",img1.shape)
print("Image -1 Type: ", img1.dtype)

print("Image -2 Shape: ",img2.shape)
print("Image -2 Type: ", img2.dtype)

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# check image shape and channel
print("Image -1 Shape post conversion: ",img1.shape)
print("Image -1 Type post conversion: ", img1.dtype)

print("Image -2 Shape post conversion: ",img2.shape)
print("Image -2 Type post conversion: ", img2.dtype)

# show image with matplotlib
plt.subplot(121),plt.imshow(img1, cmap=cm.gray, vmin=0, vmax=255),plt.title('IMAGE-1'),plt.axis("Off")
plt.subplot(122),plt.imshow(img2, cmap=cm.gray, vmin=0, vmax=255),plt.title('IMAGE-1'),plt.axis("Off")
```

# Arithmatic operation on Images
Several arithmetic operations on images like addition, subtraction, bitwise operations etc.cv2.add(),cv2.addWeighted() ,etc.

This includes bitwise AND, OR, NOT and XOR operations. They will be highly useful while extracting any part of the image (as we will see in next notebook), defining and working with non-rectangular ROI etc. 

```python
# use cv2.absdiff function to find the difference
img3 = cv2.absdiff(img1,img2)
plt.imshow(img3, cmap=cm.gray, vmin=0, vmax=255)
```
